```python
# Scientific Calculator using Tkinter

import tkinter as tk
import math

class Calculator:
    def __init__(self, master):
        self.master = master
        master.title("Scientific Calculator")

        self.result = tk.StringVar()
        self.entry = tk.Entry(master, textvariable=self.result, width=40, borderwidth=10)
        self.entry.grid(row=0, column=0, columnspan=10)

        self.create_buttons()

    def create_buttons(self):
        buttons = [
            '7', '8', '9', '/',
            '4', '5', '6', '*',
            '1', '2', '3', '-',
            '0', '.', '=', '+',
            'sin', 'cos', 'tan', 'sqrt',
            '(', ')', 'C', 'log'
        ]

        row_val = 1
        col_val = 0

        for button in buttons:
            tk.Button(self.master, text=button, width=10, height=2, command=lambda b=button: self.on_button_click(b)).grid(row=row_val, column=col_val)
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

    def on_button_click(self, button):
        if button == '=':
            try:
                expression = self.result.get()
                if 'sin' in expression:
                    expression = expression.replace('sin', 'math.sin')
                elif 'cos' in expression:
                    expression = expression.replace('cos', 'math.cos')
                elif 'tan' in expression:
                    expression = expression.replace('tan', 'math.tan')
                elif 'sqrt' in expression:
                    expression = expression.replace('sqrt', 'math.sqrt')
                elif 'log' in expression:
                    expression = expression.replace('log', 'math.log')
                self.result.set(eval(expression))
            except Exception as e:
                self.result.set("Error")
        elif button == 'C':
            self.result.set("")
        else:
            current_text = self.result.get()
            self.result.set(current_text + button)

if __name__ == "__main__":
    root = tk.Tk()
    calculator = Calculator(root)
    root.mainloop()
```

```python
import tkinter as tk
import math

class Calculator:
    def __init__(self, master):
        self.master = master
        master.title("Scientific Calculator")
        master.configure(bg="#2e2e2e")

        self.result = tk.StringVar()

        self.entry = tk.Entry(master, textvariable=self.result, font=("Helvetica", 20), borderwidth=0, relief="flat", bg="#393939", fg="#ffffff")
        self.entry.grid(row=0, column=0, columnspan=5, padx=10, pady=20, ipady=10)

        self.create_buttons()

    def create_buttons(self):
        button_texts = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
            ('sin', 5, 0), ('cos', 5, 1), ('tan', 5, 2), ('sqrt', 5, 3),
            ('(', 6, 0), (')', 6, 1), ('C', 6, 2), ('log', 6, 3),
        ]

        for (text, row, col) in button_texts:
            button = tk.Button(self.master, text=text, font=("Helvetica", 15), borderwidth=0, relief="flat", bg="#4e4e4e", fg="#ffffff", 
                               command=lambda b=text: self.on_button_click(b))
            button.grid(row=row, column=col, padx=5, pady=5, ipadx=10, ipady=10)

    def on_button_click(self, button):
        current_text = self.result.get()

        if button == '=':
            try:
                expression = self.result.get()
                expression = self.replace_math_functions(expression)
                self.result.set(eval(expression))
            except Exception as e:
                self.result.set("Error")
        elif button == 'C':
            self.result.set("")
        else:
            self.result.set(current_text + button)

    def replace_math_functions(self, expression):
        expression = expression.replace('sin', 'math.sin')
        expression = expression.replace('cos', 'math.cos')
        expression = expression.replace('tan', 'math.tan')
        expression = expression.replace('sqrt', 'math.sqrt')
        expression = expression.replace('log', 'math.log')
        return expression

if __name__ == "__main__":
    root = tk.Tk()
    calculator = Calculator(root)
    root.mainloop()
```
